```python
### import numpy as np
import matplotlib.pyplot as plt

def ReLU(x):
   return 0 if max(x) < 0 else max(x)

def PReLU(x, a):
    return ReLU(x) + (a * min(x) if min(x) <= 0 else 0)

def SoftPlus(x):
    return np.log(1 + np.exp(x))
    
x_range = np.linspace(-10,10,21)
fig, ax = plt.subplots(3,1, figsize=[6,6])

y = np.zeros((3, 21))
for i in range(21):
    y[0][i] = ReLU([x_range[i]])
    y[1][i] = PReLU([x_range[i]], 0.1)
y[2] = SoftPlus(x_range)

ax[0].plot(x_range, y[0], label="ReLU function")
ax[0].legend()

ax[1].plot(x_range, y[1], label="PReLU function | a=0.1")
ax[1].legend()

ax[2].plot(x_range, y[2], label="SoftPlus function | a=0.1")
ax[2].legend()
```
