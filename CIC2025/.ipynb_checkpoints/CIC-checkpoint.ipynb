```python
# PESQUISA ZETA
# Projeto: Anlise de Estabilidade de Solucionadores de Equaes Diferenciais Baseados em Redes Neurais (Neural ODEs)

import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
from torchdiffeq import odeint
from scipy.integrate import solve_ivp

# ====================================================
# 1. Definio do Sistema Dinmico - Oscilador Amortecido
# ====================================================
 = 0.1        # coeficiente de amortecimento
_n = 2.0      # frequncia natural
t_max = 20.0
t_eval = np.linspace(0, t_max, 200)

def true_dynamics(t, x):
    """ Equaes do oscilador amortecido (sistema de 1a ordem). """
    dx1 = x[1]
    dx2 = -2 *  * _n * x[1] - _n**2 * x[0]
    return [dx1, dx2]

# Condies iniciais
x0 = [1.0, 0.0]  # deslocamento inicial, velocidade inicial

# Soluo "ground truth" via RK45 (alta preciso)
sol = solve_ivp(true_dynamics, [0, t_max], x0, t_eval=t_eval, method='RK45', rtol=1e-10, atol=1e-12)
ground_truth = sol.y.T
t_train = torch.tensor(t_eval, dtype=torch.float32)

# ====================================================
# 2. Definio da Rede Neural para f(x)
# ====================================================
class ODEFunc(nn.Module):
    def __init__(self):
        super(ODEFunc, self).__init__()
        self.net = nn.Sequential(
            nn.Linear(2, 32),
            nn.Tanh(),
            nn.Linear(32, 2)
        )
    def forward(self, t, x):
        return self.net(x)

func = ODEFunc()

# ====================================================
# 3. Treinamento do Neural ODE
# ====================================================
optimizer = optim.Adam(func.parameters(), lr=0.0005)
loss_fn = nn.MSELoss()

# Dados de treinamento
x_train = torch.tensor(ground_truth, dtype=torch.float32)

for epoch in range(2000):
    optimizer.zero_grad()
    # Resolver ODE neural
    pred_x = odeint(func, torch.tensor(x0, dtype=torch.float32), t_train)
    loss = loss_fn(pred_x, x_train)
    loss.backward()
    optimizer.step()
    if epoch % 200 == 0:
        print(f"Epoch {epoch}, Loss: {loss.item():.6f}")

# ====================================================
# 4. Teste de Estabilidade
# ====================================================
# Perturbao inicial
eps = np.array([0.01, -0.01])
x0_perturb = [x0[0] + eps[0], x0[1] + eps[1]]

# Soluo ground truth perturbada
sol_perturb = solve_ivp(true_dynamics, [0, t_max], x0_perturb, t_eval=t_eval, method='RK45', rtol=1e-10, atol=1e-12)
ground_truth_perturb = sol_perturb.y.T

# Neural ODE original
pred_x = odeint(func, torch.tensor(x0, dtype=torch.float32), t_train).detach().numpy()
# Neural ODE perturbado
pred_x_perturb = odeint(func, torch.tensor(x0_perturb, dtype=torch.float32), t_train).detach().numpy()

# Clculo das normas de erro ||x - x_p||
def error_norm(sol1, sol2):
    return np.linalg.norm(sol1 - sol2, axis=1)

error_true = error_norm(ground_truth, ground_truth_perturb)
error_neural = error_norm(pred_x, pred_x_perturb)

# ====================================================
# 5. Visualizaes
# ====================================================
plt.figure(figsize=(12, 5))

# (a) Diagramas de fase
plt.subplot(1, 2, 1)
plt.plot(ground_truth[:, 0], ground_truth[:, 1], 'k-', label="Ground Truth (RK45)")
plt.plot(pred_x[:, 0], pred_x[:, 1], 'r--', label="Neural ODE")
plt.xlabel("x1 (posio)")
plt.ylabel("x2 (velocidade)")
plt.title("Diagrama de Fase")
plt.legend()

# (b) Normas de erro
plt.subplot(1, 2, 2)
plt.plot(t_eval, error_true, 'k-', label="Erro Ground Truth (perturbado)")
plt.plot(t_eval, error_neural, 'r--', label="Erro Neural ODE (perturbado)")
plt.xlabel("Tempo")
plt.ylabel("||x - x_p||")
plt.title("Anlise de Estabilidade")
plt.legend()

plt.tight_layout()
plt.savefig("ep2/cic13.png")
plt.show()
```
