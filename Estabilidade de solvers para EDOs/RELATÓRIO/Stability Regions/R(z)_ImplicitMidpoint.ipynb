# Região de Estabilidade do Método do Ponto Médio Implícito

Para o método do ponto médio implícito, usamos a fórmula transformada na versão RK dada pela equação:
$$
    u_{n+1} = u_{n} + \Delta t \ f\left( t_{n} + \Delta t, \ \frac{u_{n} + u_{n+1}}{2} \right)
$$

substituindo pela equação de teste de Dahlquist, obtemos:
$$
    u_{n+1} = u_{n} + \Delta t \ \lambda \left( \frac{u_{n} + u_{n+1}}{2} \right) \\
$$

como $z = \Delta t \lambda$, podemos reescrever como:
$$
    \begin{align}
    u_{n+1} = u_{n} + z\left( \frac{u_{n}  + u_{n+1}}{2} \right)  \\
    u_{n+1} = u_{n} + \frac{ zu_{n}}{2}  + \frac{zu_{n+1}}{2}   \\
    2u_{n+1} = 2u_{n} + zu_{n} + zu_{n+1} 
    \end{align}
$$

agora vamos manipular os termos:
$$
    \begin{align}
    u_{n+1} (2 -z) &= u_{n}(2 +z) \\
    u_{n+1} &= \frac{u_{n}(2+z)}{2-z} \\
    \end{align}
$$

assumindo a definição de $R(z)$, encontramos a região de estabilidade:
$$
    R(z) = \frac{2 + z}{2-z}
$$

```python
import numpy as np 
import matplotlib.pyplot as plt 

x = np.linspace(-3.0, 3, 400)
y = np.linspace(-3.0, 3.0, 400)
xx, yy = np.meshgrid(x, y)
z = xx + 1j*yy

def ImplictMidpoint(z):
    return (2 + z) / (2 - z)


RE_IMP = np.abs(ImplictMidpoint(z))


plt.figure(figsize=(10,10))
plt.title('Ponto Médio Implícito ($|R(z)| ≤ 1$)')
plt.contourf(xx, yy, RE_IMP, levels=[0,1], colors='lightgray')
plt.contour(xx, yy, RE_IMP, levels=[1], colors="gray")
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')
plt.grid(True)
plt.axhline(0, color='black', lw=1, zorder=2) 
plt.axvline(0, color='black', lw=1, zorder=2)
```
