# Método de Euler Implícito (Backward Euler) - BE

Primeiro para o método BE. Substituindo $f(t_{n+1},u_{n+1})$ por $\lambda u_{n+1}$, obtemos: 
    $$
    u_{n+1} - \Delta t \ \lambda u_{n+1} = u_{n} 
    $$
simplificando:
    $$
    u_{n+1} (1 - \Delta t \lambda) = u_{n} 
    $$

e como $z = \Delta t \lambda$ 
    $$
    \begin{align*}
    u_{n+1}(1 - z) &= u_{n}  \\
    u_{n+1} &= \frac{u_{n}}{(1-z)}
    \end{align*}
    $$

e aplicando $R(z)=\frac{u_{n+1}}{u_{n}}$, chegamos na região de estabilidade 
    $$
    R(z) = \frac{1}{(1-z)}
    $$


```python
import numpy as np 
import matplotlib.pyplot as plt 

x = np.linspace(-3.0, 3, 400)
y = np.linspace(-3.0, 3.0, 400)
xx, yy = np.meshgrid(x, y)
z = xx + 1j*yy

def BackwardEuler(z):
    return 1 / (1-z)


RE_BE = np.abs(BackwardEuler(z))


plt.figure(figsize=(10,10))
plt.title('Euler Implícito ($|R(z)| ≤ 1$)')
plt.contourf(xx, yy, RE_BE, levels=[0,1], colors='lightgray')
plt.contour(xx, yy, RE_BE, levels=[1], colors="gray")
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')
plt.grid(True)
plt.axhline(0, color='black', lw=1, zorder=2) 
plt.axvline(0, color='black', lw=1, zorder=2)
```
