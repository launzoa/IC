# Região de Estabilidade para o Método de Euler Explícito (FE - Forward Euler)

A construção teórica para a região é dada por: 
$$
    \begin{align*}
    u_{n+1} &= u_{n} + \Delta t \ f(t_{n},u_{n})  \\  
    &= u_{n} + \Delta t \ \lambda u_{n} \\
    & = u_{n} (1 + \Delta t \ \lambda) 
    \end{align*}
$$

como temos que $z = \Delta t \lambda$, podemos substituir na fórmula e obter:
$$
    \begin{align*}
    u_{n+1} = u_{n} (1 + z)
    \end{align*}
$$

e pela definição de $R(z) \cdot u_n = u_{n+1}$, chegamos a região de estabilidade: 
$$
    \begin{align*}
    R(z) = 1 + z
    \end{align*}
$$

onde sabemos que $R(z) \le 1$, logo $|1 + z| \le 1$.

indo mais a fundo, vamos encontrar as coordenadas. Primeiro substituimos $z = x + yi$
$$
    \begin{align*}
    |1 + (x + yi)| \le 1 \\
    |(x + 1) + yi| \le 1
    \end{align*}
$$

e como o módulo de um soma de número complexo é dado por $\sqrt{a^2 + b^2}$, onde $a$ é a parte real e $b$ a imaginária
$$
    \begin{align*}
    \sqrt{(x + 1)^2 + y^2} \le 1
    \end{align*} 

$$

elevando ambos os lados ao quadrado, obtemos:
$$
    \begin{align*}
    (x + 1)^2 + y^2 \le 1^2
    \end{align*} 

$$

Portanto,o resultado é um círculo de raio 1 e centro em (-1,1)


```python
import numpy as np
import matplotlib.pyplot as plt 

x = np.linspace(-3.0, 2.0, 400)
y = np.linspace(-2.5, 2.5, 400)
xx, yy = np.meshgrid(x,y)   
z = xx + 1j*yy               

def FE(z):
    return 1 + z 

RE_Euler = np.abs(FE(z))

plt.figure(figsize=(10,10))
plt.title('Euler Explícito ($|R(z)| ≤ 1$)')
plt.contourf(xx, yy, RE_Euler, levels=[0,1], colors='lightgray')
plt.contour(xx, yy, RE_Euler, levels=[1], colors="gray")
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')
plt.grid(True)
plt.axhline(0, color='black', lw=1, zorder=2) 
plt.axvline(0, color='black', lw=1, zorder=2)
```
