# Método de Heun (Método Trapezoidal Explícito)

A região de estabilidade do método de Heun é similar ao do ponto médio, temos: 
    $$
        u_{n+1} = u_{n} + \frac{\Delta t}{2}(k_{1}+k_{2})
    $$
    
e resolvendo primeiro a equação de $k_{1}$:
    $$
    \begin{align*}
        k_{1} &= f(t_{n}, u_{n})  \\ 
        k_{1} &=  \lambda u_{n} 
    \end{align*}
    $$
    
agora de $k_{2}$: 
    $$
    \begin{align*}
        k_{2} & = f(t_{n} + \Delta t, \ u_{n} + \Delta t \ k_{1})  \\
        k_{2} &= f(t_{n} + \Delta t, \ u_{n} + \Delta t \ \lambda u_{n})\\
        k_{2} &= \lambda \cdot ( u_{n }+ \Delta t  \lambda u_{n})  \\
        k_{2} &= \lambda u_{n} \cdot(1 + \Delta t \lambda)
    \end{align*}
    $$

como temos que $z = \Delta t \lambda$, reescrevemos como: 
    $$
    k_{2} = \lambda u_{n}  \cdot(1 + z)
    $$
    
agora que encontramos os valores para $k_{1}$ e $k_{2}$, substituímos na sua equação: 
    $$
    u_{n+1} = u_{n} + \frac{\Delta t}{2}(\lambda u_{n} + \lambda u_{n}(1 + z))
    $$
    
fatorando $\lambda u_{n}$, obtemos: 
    $$
    u_{n+1} = u_{n} + \frac{\Delta t\lambda \ u_{n}}{2}(1 + (1+z))
    $$
    
e como $\Delta t\lambda = z$, chegamos a fórmula resumida 
    $$
    u_{n+1} = u_{n} + \frac{zu_{n}}{2} (2 + z)
    $$
    
como a equação está em função de $u_{n}$, aplicamos o $R(z) \cdot u_n = u_{n+1} $: 
    $$
    R(z) = 1 + \frac{z}{2} (2 + z)
    $$
    
agora expandindo a fórmula, temos: 
    $$
    R(z) = 1 + z + \frac{z^2}{2}
    $$

como a sua região de estabilidade é a mesma que a do Ponto Médio, consequentemente sua interpretação geométrica será a mesma.


```python
import numpy as np 
import matplotlib.pyplot as plt 

x = np.linspace(-3.0, 2.0, 400)
y = np.linspace(-3.0, 3.0, 400)
xx, yy = np.meshgrid(x, y)
z = xx + 1j*yy

def Heun(z):
    return 1 + z + (z**2) / 2

RE_HE = np.abs(Heun(z)) 

plt.figure(figsize=(10,10))
plt.title('Heun ($|R(z)| ≤ 1$)')
plt.contourf(xx, yy, RE_HE, levels=[0,1], colors='lightgray')
plt.contour(xx, yy, RE_HE, levels=[1], colors="gray")
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')
plt.grid(True)
plt.axhline(0, color='black', lw=1, zorder=2) 
plt.axvline(0, color='black', lw=1, zorder=2)
```
