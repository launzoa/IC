Inicialização dos valores e parâmetros

```python
import torch.nn as nn
import torch
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns 

age = np.array([0, 0.1, 0.2, 0.25, 0.3, 0.4, 0.5, 0.7, 0.9])
value = np.array([((x-0.3)**2 - 0.45*x  + 0.3) * 2 + np.random.randn() / 60 for x in age])

df = pd.DataFrame({"Age": age, "Value":value})

x = torch.FloatTensor([df.Age.values.astype(float)])
y = torch.FloatTensor(df.Value.values.astype(float))
y = y.unsqueeze(1)
x = x.view(x.size(1), -1)
```

Rede Neural

```python
class Model(nn.Module):
    def __init__(self, hidden_neurons=2):
        super().__init__()

        self.hidden_neurons = hidden_neurons
        self.input_layer = nn.Linear(in_features=1, out_features=self.hidden_neurons, bias=True)
        self.output_layer = nn.Linear(in_features=2, out_features=1)
        self.activation = nn.Sigmoid()

    def forward(self, x):
        out = self.input_layer(x)
        out = self.activation(out)
        out = self.output_layer(out)
        return out

EPOCHS = 1000
LR = 0.1

model = Model()
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(params=model.parameters(), lr=LR)

weights = []
bias = []

for epoch in range(EPOCHS):
    outputs = model.forward(x)

    loss = criterion(outputs, y)

    optimizer.zero_grad()

    loss.backward()

    optimizer.step()

    weights.append(model.input_layer.weight)
    bias.append(model.input_layer.bias)
    
```

Impressão

```python
out = outputs

print(f"Input Layer:\nPesos: {model.input_layer.weight.view(-1).numpy(force=True)}\nBias: {model.input_layer.bias.view(-1).numpy(force=True)}")
print(f"\nHidden Layer:\nPesos: {model.output_layer.weight.view(-1).numpy(force=True)}\nBias: {model.output_layer.bias.view(-1).numpy(force=True)}")
```

Modelagem dos parâmetros de entrada

```python
def tensor_to_numpy(tensor: torch.tensor):
    return tensor.view(-1).numpy(force=True)
    
x_range = torch.FloatTensor(np.arange(0, 1.0, 0.02))

y1 = x_range*model.input_layer.weight[0][0] + model.input_layer.bias[0]
sigmoid = 1 / (1 + torch.exp(-y1))
output = sigmoid*model.input_layer.weight[0][0]

y1_numpy = tensor_to_numpy(y1)
sigmoid_numpy = tensor_to_numpy(sigmoid)
output_numpy = tensor_to_numpy(output)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=[15,5])

ax[0].plot(x_range, y1_numpy, lw=3)
ax[0].set_xlabel('Age')
ax[0].set_ylabel('y')
ax[0].set_title("Somatório")

ax[1].plot(x_range, sigmoid_numpy)
ax[1].set_xlabel('Age')
ax[1].set_ylabel('y')
ax[1].set_title("Função de Ativação (Sigmoid)")

ax[2].plot(x_range, output_numpy)
ax[2].set_xlabel('Age')
ax[2].set_ylabel('y')
ax[2].set_title("Saída")

plt.tight_layout(pad=0.5)
```

Modelagem dos parâmetros de saída

```python
y2 = x_range*model.input_layer.weight[1][0] + model.input_layer.bias[1]
sigmoid2 = 1 / (1 + torch.exp(-y2))
output2 = sigmoid2*model.input_layer.weight[0][0]

y2_numpy = tensor_to_numpy(y2)
sigmoid_numpy2 = tensor_to_numpy(sigmoid2)
output_numpy2 = tensor_to_numpy(output2)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=[15,5])

ax[0].plot(x_range, y2_numpy, lw=3)
ax[0].set_xlabel('Age')
ax[0].set_ylabel('y')
ax[0].set_title("Somatório")

ax[1].plot(x_range, sigmoid_numpy2)
ax[1].set_xlabel('Age')
ax[1].set_ylabel('y')
ax[1].set_title("Função de Ativação (Sigmoid)")

ax[2].plot(x_range, output_numpy2)
ax[2].set_xlabel('Age')
ax[2].set_ylabel('y')
ax[2].set_title("Saída")

plt.tight_layout(pad=0.5)
```

Saída final, soma do treinamento de entrada mais o treinamento de saída

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=[15,5])

ax.plot(x_range, output_numpy, label="Neurônio 1")
ax.plot(x_range, output_numpy2, label="Neurônio 2")
ax.plot(x_range, (output_numpy + output_numpy2), label="Neurônio 1 + 2")
ax.set_xlabel('Age')
ax.set_ylabel('y')
ax.legend()

plt.tight_layout(pad=0.5)
```

```python
out_bias = output + output2 + model.output_layer.bias[0]
out_bias_numpy = tensor_to_numpy(out_bias)

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=[15,5])

ax.plot(x_range, (output_numpy + output_numpy2), label="Neurônio 1 + 2")
ax.plot(x_range, out_bias_numpy, label="Neurônio 1 + 2 + Bias")
ax.set_xlabel('Age')
ax.set_ylabel('y')
ax.legend()

plt.tight_layout(pad=0.5)
```
