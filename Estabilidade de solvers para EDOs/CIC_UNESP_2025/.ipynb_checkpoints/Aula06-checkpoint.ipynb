```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):

    return 1 / (1 +np.exp(-x) )
    
def tanh(x):
    return ( np.exp(x) - np.exp(-x) ) / ( np.exp(x) + np.exp(-x) ) 

x_range = np.arange(-6,6,0.1)


fig, ax = plt.subplots(1,2, figsize=[10,4])
ax[0].plot(x_range, sigmoid(x_range), label='Sigmoid function')
ax[1].plot(x_range, tanh(x_range), label='Tangent Hiperbolic function')
```
