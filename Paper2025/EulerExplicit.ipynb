```python
import numpy as np
import matplotlib.pyplot as plt 

fig, ax = plt.subplots(figsize=(7, 7))

centro_x = -1
centro_y = 0
raio = 1

circulo = plt.Circle((centro_x, centro_y), raio, 
                     facecolor='orange', edgecolor='black', alpha=0.5,
                     label='Região |z + 1| < 1')

ax.add_patch(circulo)

ax.axhline(0, color='black', linewidth=0.8)
ax.axvline(0, color='black', linewidth=0.8)

ax.set_aspect('equal', adjustable='box')

ax.grid(True, linestyle='--', alpha=0.6)

ax.set_xlim(-2.5, 1.5)
ax.set_ylim(-2, 2)

ax.set_title('Região de Estabilidade do Método de Euler Explícito', fontsize=10)
ax.set_xlabel('Eixo Real (x)', fontsize=12)
ax.set_ylabel('Eixo Imaginário (y)', fontsize=12)
ax.legend()

plt.show()
```

```python
f = lambda x, y: -20 * y
exact_solution = lambda x, k: k * np.exp(-20 * x)

a = 0; b = 0.5
h_range = [0.1, 0.05, 0.025]
k = 1



for h in h_range:
    n = int((b - a) / h)
    t = np.arange(a,b + h, h)

    w = np.zeros(n+1)
    w[0] = k
    
    for i in range(n):
        w[i+1] = w[i] + h * f(t[i], w[i])

    plt.plot(t, w, label="Euler h")

y = exact_solution(t, k)

plt.plot(t, y, label="Solução exata ($k * e^{-20x}$)")
plt.legend()
plt.show()
```
